```python
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

```python
dataset2 = randn(80)
```

```python
plt.hist(dataset2, color='indianred')
```

```python
plt.hist(dataset1, density=True)
```

```python
plt.hist(dataset1, density=True, alpha=0.5, bins=20)
plt.hist(dataset2, density=True, alpha=0.5, bins=20)
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
sns.jointplot(data1, data2)
```

```python
sns.jointplot(data1, data2, kind='hex')

```
